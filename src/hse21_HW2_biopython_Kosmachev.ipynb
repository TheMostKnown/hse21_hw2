## Добавляем координаты генов

```python
!head -n 20 proteins.fasta
```

Можем заметить, что тут то же самое, что и в файле *gms2.lst*, однако это легче обрабатывать

```python
from pprint import pprint
from Bio.SeqFeature import SeqFeature, FeatureLocation

scaf_genes = {}

for gene in SeqIO.parse("proteins.fasta", "fasta"):
  description = list(map(str, gene.description.split()))
  id = description[0]
  scaffold = description[1]
  start = int(description[2])
  stop = int(description[3])
  strand = description[4]
  if strand == '-':
    strand = -1
  else:
    strand = 1

  feature = SeqFeature(FeatureLocation(start, stop, strand=strand), type="CDS")
  feature.qualifiers['locus_tag'] = [id]
  feature.qualifiers['translation'] = [gene.seq]

  if scaf_genes.get(scaffold) == None:
    scaf_genes[scaffold] = [feature]
  else:
    scaf_genes[scaffold].append(feature)
```

```python
pprint(scaf_genes['scaffold10_len158_cov303_single'][0])
```

```python
none_count = 0
for record in record_list:
  if scaf_genes.get(record.id) == None:
    none_count += 1
  else:
    record.features = scaf_genes.get(record.id)
  print(record)
```

```python
none_count
```

```python
SeqIO.write(record_list, "GENOME.gbk", "genbank")
```

## Добавляем функции белков (из бактерии MIL-1)

```python
!head -n 20 scaffolds.hits_from_MIL_1.txt
```

```python
id_ccu = {}

with open ('scaffolds.hits_from_MIL_1.txt', 'r') as hits:
  hit = hits.readline()
  while (hit != ''):
    data = list(map(str, hit.split('\t')))
    if (data[1].find('_CCU') != -1):
      id = data[0]
      info = list(map(str, data[1].split('_')))
      info = info[2]
      if (id_ccu.get(id) == None):
        id_ccu[id] = [info]
      else:
        id_ccu[id].append(info)
    hit = hits.readline()
hits.close()
```

```python
for elem in id_ccu:
  id_ccu[elem] = set(id_ccu[elem])
  id_ccu[elem] = list(id_ccu[elem])
```

```python
print(id_ccu['7'])
```

```python
for gene_value_list in scaf_genes.values():
  for gene_value in gene_value_list:
    if id_ccu.get(gene_value.qualifiers['locus_tag'][0]) != None:
      gene_value.qualifiers['note'] = id_ccu.get(gene_value.qualifiers['locus_tag'][0])
```

```python
pprint(scaf_genes['scaffold11_len2560_cov340'][0].qualifiers) # Последняя строка появилась, потому что немного улучшал отображение
```

```python
mil_1_genome = SeqIO.read("/content/T_oleivorans_MIL_1.gbk", "genbank")
```

```python
for  mil_f in mil_1_genome.features:
  if 'protein_id' not in mil_f.qualifiers:
    continue
  
  for feature_value_list in scaf_genes.values():
    for feature_value in feature_value_list:
      if feature_value.qualifiers.get('note') != None:
        for i in range(len(feature_value.qualifiers['note'])):
          if feature_value.qualifiers['note'][i] == mil_f.qualifiers['protein_id'][0]:
            if (feature_value.qualifiers.get('product') != None):
              feature_value.qualifiers['product'].append(mil_f.qualifiers['product'])
            else:
              feature_value.qualifiers['product'] = mil_f.qualifiers['product']
      else: feature_value.qualifiers['product'] = None
```

```python
for feature_value_list in scaf_genes.values():
    for feature_value in feature_value_list:
      if feature_value.qualifiers.get('product') != None:
        tmp = set()
        for product in feature_value.qualifiers['product']:
          tmp.add(product[0])
        feature_value.qualifiers['product'] = list(tmp)
      pprint(feature_value.qualifiers)
```

```python
for record in record_list:
  if scaf_genes.get(record.id) != None:
    record.features = scaf_genes[record.id]
  print(record)
SeqIO.write(record_list, "GENOME.gbk", "genbank")
```
